```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("time_series_s.csv", index_col="timestamp", parse_dates=True)
df = df.iloc[::-1]
df
```

### SMA

```python
def sma(data):
    n = int(input("SMA period: "))
    return data["close"].rolling(n).mean()
```

```python
df["sma"] = sma(df)
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["close"], label="Price")
plt.plot(df["sma"], label="SMA")
plt.legend()
plt.show()
```

### ADR

```python
from itertools import accumulate
```

```python
def adr(data):
    money_flow_multiplier = ((data["close"]-data["low"])-(data["high"]-data["close"]))/(data["high"]-data["low"])
    money_flow_volume = money_flow_multiplier * data["volume"]
    adr = accumulate(money_flow_volume)
    return list(adr)
```

```python
df["adr"] = adr(df)
df
```

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(211)  # Params: row, column, cell
ax2 = fig.add_subplot(212)

ax1.plot(df["close"], label="Price")
ax2.plot(df["adr"], color="orange", label="ADR")
plt.legend(loc="best")
plt.show()
```
